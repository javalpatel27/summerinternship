```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

date_columns = [
    'Date_of_Infection', 'Hospital_Admission_Date', 'Hospital_Discharge_Date',
    'Date_of_Recovery', 'Date_of_Reinfection', 'Date_of_Last_Dose'
]

for col in date_columns:
    df[col] = pd.to_datetime(df[col], errors='coerce')

for col in ['Date_of_Recovery', 'Date_of_Last_Dose']:
    df[col].fillna(pd.Timestamp('1900-01-01'))


plt.figure(figsize=(14, 6))
sns.histplot(df['Age'], kde=True, bins=30, color='pink')
plt.title('Age Distribution of Patients')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
sns.set(style="whitegrid")

plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Gender', hue='Gender', palette='pastel', legend=False)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 6))
recovery_counts = df['Recovered'].value_counts()
plt.pie(recovery_counts, labels=recovery_counts.index, autopct='%1.1f%%', colors=sns.color_palette('pastel'))
plt.title('Recovery Status')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Age', y='BMI', hue='Severity', palette='coolwarm', alpha=0.7)
plt.title('Age vs BMI by Severity')
plt.xlabel('Age')
plt.ylabel('BMI')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(df['BMI'], bins=30, kde=True, color='coral')
plt.title('BMI Distribution')
plt.xlabel('BMI')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
hospitalized_df = df[df['Hospitalized'] == 'Yes'].copy()
hospitalized_df['Admission_Month'] = hospitalized_df['Hospital_Admission_Date'].dt.to_period('M').astype(str)
monthly_hospitalizations = hospitalized_df['Admission_Month'].value_counts().sort_index()

plt.figure(figsize=(12, 5))
sns.lineplot(x=monthly_hospitalizations.index, y=monthly_hospitalizations.values, marker='o', color='green')
plt.title('Monthly Hospitalizations Over Time')
plt.xlabel('Month')
plt.ylabel('Number of Hospitalizations')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
df['Date_of_Infection'] = pd.to_datetime(df['Date_of_Infection'], errors='coerce')

# Drop missing dates if necessary
df = df.dropna(subset=['Date_of_Infection'])

# Group by date and count number of infections
daily_counts = df.groupby('Date_of_Infection').size()

# Plot time series
plt.figure(figsize=(12, 6))
sns.lineplot(x=daily_counts.index, y=daily_counts.values, marker='o', color='steelblue')
plt.title('Daily COVID Infections Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Infections')
plt.xticks(rotation=50)
plt.tight_layout()
plt.grid(True)
plt.show()
```
