```python
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].fillna('Unknown')
for col in df.select_dtypes(include='number').columns:
    df[col] = df[col].fillna(df[col].median())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 5))
sns.histplot(df['normalized_salary'], kde=True, color='skyblue')
plt.title('Distribution of Normalized Salary')
plt.xlabel('Normalized Salary')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
top_titles = df['title'].value_counts().nlargest(10)
top_titles_df = top_titles.reset_index()
top_titles_df.columns = ['title', 'count']
plt.figure(figsize=(10, 6))
sns.barplot(data=top_titles_df, x='count', y='title', hue='title', palette='pastel', dodge=False, legend=False)
plt.title('Top 10 Job Titles')
plt.xlabel('Count')
plt.ylabel('Title')
plt.show()
```

```python
plt.figure(figsize=(7, 5))
sns.scatterplot(data=df, x='views', y='normalized_salary', hue='formatted_experience_level', palette='Set1')
plt.title('Salary vs Job Views by Experience Level')
plt.xlabel('Views')
plt.ylabel('Normalized Salary')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
columns_to_plot = ['views', 'applies', 'normalized_salary']
df_sorted = df.sort_values(by='listed_time')
for col in columns_to_plot:
    plt.plot(df_sorted['listed_time'], df_sorted[col], label=col)

plt.title('Trends Over Time')
plt.xlabel('Listed Time')
plt.ylabel('Value')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.regplot(data=df, x='views', y='normalized_salary', lowess=True, scatter_kws={'s': 10}, line_kws={'color': 'red'})
plt.title('Smoothed Trend: Views vs Salary')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
df_3d = df[['normalized_salary', 'views', 'applies']].dropna()
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(
    df_3d['normalized_salary'],
    df_3d['views'],
    df_3d['applies'],
    c=df_3d['views'], cmap='viridis', alpha=0.7
)
ax.set_xlabel('Normalized Salary')
ax.set_ylabel('Views')
ax.set_zlabel('Applies')
ax.set_title('3D Scatter Plot: Salary vs Views vs Applies')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
work_type_counts = df['work_type'].value_counts().nlargest(5)
plt.figure(figsize=(4, 4))
plt.pie(work_type_counts.values, labels=work_type_counts.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Pastel1.colors)
plt.title('Top 5 Work Types Distribution')
plt.axis('equal')  # Equal aspect ratio ensures the pie is circular
plt.show()
```

```python
df['listed_time'] = pd.to_datetime(df['listed_time'], errors='coerce')
df_sorted = df.sort_values('listed_time').dropna(subset=['views', 'applies', 'normalized_salary'])
df_sorted['timestamp'] = df_sorted['listed_time'].astype('int64') // 10**9  
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot(
    df_sorted['timestamp'],
    df_sorted['views'],
    df_sorted['applies'],
    color='royalblue'
)
ax.set_xlabel('Time (Unix)')
ax.set_ylabel('Views')
ax.set_zlabel('Applies')
ax.set_title('3D Line Plot Over Time')

plt.tight_layout()
plt.show()
```

```python
top_companies = df['company_name'].value_counts().nlargest(5)

plt.figure(figsize=(4, 4))
plt.pie(top_companies.values, labels=top_companies.index, autopct='%1.1f%%', startangle=120, colors=plt.cm.Set2.colors)
plt.title('Top 5 Companies by Number of Postings')
plt.axis('equal')
plt.show()
```
